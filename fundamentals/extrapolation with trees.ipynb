```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(42)
n_samples = 100
x = np.linspace(0, 100, n_samples)
y = 2 * np.sin(x) - 0.5 * x + 10 + np.random.normal(0, 0.5, n_samples)

# Reshape the input data
x = x.reshape(-1, 1)

# Create and fit the decision tree regressor
regressor = DecisionTreeRegressor()
regressor.fit(x, y)

# Generate predictions
x_test = np.linspace(min(x), max(x), n_samples).reshape(-1, 1)
y_pred = regressor.predict(x_test)

# Plot the results
plt.scatter(x, y, label='Training Data')
plt.plot(x_test, y_pred, color='r', label='Decision Tree Regression')

xOutsideLearning= np.array([150,200]).reshape(-1, 1)
plt.scatter(xOutsideLearning,regressor.predict(xOutsideLearning), label='Data outside the training set')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Decision Tree Regression')
plt.legend()
plt.show()
```
