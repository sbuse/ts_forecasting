```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sdf = pd.read_csv("data/swissGridExport.csv", names=["timestamp", "dailySumkWh"])                 
sdf['timestamp'] = pd.to_datetime(sdf['timestamp'])

#sort the dataframe by timestamp
sdf = sdf.sort_values(by=['timestamp'])

#remove first row
sdf = sdf.iloc[1:]
sdf.index = sdf.timestamp
sdf = sdf.drop(columns=['timestamp'])
sdf = sdf[sdf.index < '2020-01-01']
```

```python
plt.figure(figsize=(15, 3))
plt.plot(sdf)
```
